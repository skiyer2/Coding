Credit Card analysis of customers:

This is a sample credit card dataset case which requires to develop a customer segmentation to define marketing strategy. The
sample Dataset summarizes the usage behaviour of about 9000 active credit card holders during the 6 months period. The file is at a customer level with 18 behavioural features.

Please refer to https://github.com/Accelerate-AI/UnsupervisedLearning_Clustering/blob/main/DailyQuestions/CreditCard.csv for dataset.

Following is the Data Dictionary for Credit Card dataset :-

- CUSTID : Identification of Credit Card holder
- BALANCE : Balance amount left in their account to make purchases
- BALANCE_FREQUENCY : How frequently the Balance is updated, score between 0 and 1 (1 = frequently updated, 0 = not frequently updated)
- PURCHASES : Amount of purchases made from account
- ONEOFF_PURCHASES : Maximum purchase amount done in one-go
- INSTALLMENTS_PURCHASES : Amount of purchase done in installment
- CASH_ADVANCE : Cash in advance given by the user
- PURCHASES_FREQUENCY : How frequently the Purchases are being made, score between 0 and 1 (1 = frequently purchased, 0 = not frequently purchased)
- ONEOFF_PURCHASES_FREQUENCY : How frequently Purchases are happening in one-go (1 = frequently purchased, 0 = not frequently purchased)
- PURCHASES_INSTALLMENTS_FREQUENCY : How frequently purchases in installments are being done (1 = frequently done, 0 = not frequently done)
- CASH_ADVANCE_FREQUENCY : How frequently the cash in advance being paid
- CASH_ADVANCE_TRX : Number of Transactions made with "Cash in Advanced"
- PURCHASES_TRX : Number of purchase transactions made
- CREDIT_LIMIT : Limit of Credit Card for user
- PAYMENTS : Amount of Payment done by user
- MINIMUM_PAYMENTS : Minimum amount of payments made by user
- PRC_FULL_PAYMENT : Percent of full payment paid by user
- TENURE : Tenure of credit card service for user

Ask:
- In this above dataset, what is the optimum number of clusters?

- Compute inertia values for K=3,4,5 and what do you infer from them?

- Do you see any difference in their cluster centers (3 cases with K=3,4,5)?

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
# Load data
df = pd.read_csv('CreditCard.csv')
df.head()
```

```python
# Drop CUST_ID
df.drop('CUST_ID', axis=1, inplace=True)
```

```python
# Check data types
df.info()
```

```python
# Check missing values
(df.isna().sum()/df.shape[0])*100
```

```python
# Check MINIMUM_PAYMENTS
df[df['MINIMUM_PAYMENTS'].isna()]
```

```python
# Check distribution
sns.displot(data=df, x='MINIMUM_PAYMENTS')
plt.show()
```

```python
# Lets impute MINIMUM_PAYMENTS with 0
df['MINIMUM_PAYMENTS'].fillna(0, inplace=True)
```

```python
# Check missing values again
df.isna().sum()
```

```python
# Check CREDIT_LIMIT
df[df['CREDIT_LIMIT'].isna()]
```

```python
# Lets drop the above record
df.dropna(inplace=True)
```

```python
# Check shape
df.shape
```

```python
# Find optimum clusters
K = [3, 4, 5]
results = []

for c in K:
    clus = KMeans(n_clusters=c, random_state=34, init='k-means++', n_init=100)
    clus_labels = clus.fit_predict(df)
    silhoutte_avg = silhouette_score(df, clus_labels)
    results.append([c, silhoutte_avg])
    
res = pd.DataFrame(results, columns=['n_clusters', 'silhoutte_score'])
pivot_km = pd.pivot_table(res, index='n_clusters', values='silhoutte_score')

plt.figure()
sns.heatmap(pivot_km, annot=True, linewidths=0.5, fmt='.3f', cmap=sns.cm.rocket_r)
plt.tight_layout()
```

silhoutte score depicts 4 clusters can be considered as optimal one.

```python
# Check cluster centers
K = [3, 4, 5]

for k in K:
    km = KMeans(n_clusters=k, n_init=100, init='k-means++', random_state=34)
    km = km.fit(df)
    c = km.cluster_centers_
    print('no. of clusters:', k)
    print(c)
    print()
```

cluster centers are different for each case for K = 3,4 and 5

```python
# Relationship between clsuter no. and Inertia

K = [3, 4, 5]

inertias = []

for k in K:
    kmeansmodel = KMeans(n_clusters=k, n_init=100, init='k-means++', random_state=34).fit(df)
    inertias.append(kmeansmodel.inertia_)

# Plot
plt.plot(K, inertias, 'bx-')
plt.xlabel('Value of K')
plt.ylabel('Inertia')
plt.title('Relationship between clsuter no. and Inertia')
plt.show()
```

As the number of clusters increase, inertia decreases.
