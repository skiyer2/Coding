Explain the relationship between cluster number and inertia value? (i.e. if no of cluster increases, then inertia increases or decreases for Kmeans)

Inertia value in KMeans: It is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster. Ideally, a good model is one with low inertia AND a low number of clusters (K).

````python
# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
````

````python
# Data
X, y = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=0.5, random_state=3)
````

````python
# Check X
X
````

````python
# Check y
y
````

````python
# Relationship between clsuter no. and Inertia

K = range(1, 5)

inertias = []

for k in K:
    kmeansmodel = KMeans(n_clusters=k).fit(X)
    inertias.append(kmeansmodel.inertia_)

# Plot
plt.plot(K, inertias, 'bx-')
plt.xlabel('Value of K')
plt.ylabel('Inertia')
plt.title('Relationship between clsuter no. and Inertia')
plt.show()
````

Inference:<br>
As the number of clsuters increases the inertia is decreasing. For cluster 1 (corresponding to cluster number 0), the inertia was the highest and it decreased drastically for cluster 2 (corresponding to cluster number 1).

A good model is the one which has low inertia and low number of clusters. Hence cluster 2 (corresponding to cluster number 1) can be chosen which means the optimum clusters for this model is 2.
