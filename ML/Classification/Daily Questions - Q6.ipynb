```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
from statsmodels.api import Logit, add_constant
import statsmodels.formula.api as smf

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data
df = pd.read_csv('C:/Users/Karthik.Iyer/Downloads/AccelerateAI/Classification-Models-main/data/InsuranceCrossSell.csv')
df.head()
```

```python
# Lets fit the Logit without converting categorical variables
f = 'Response ~ C(Gender) + Age + Driving_License + Region_Code + Previously_Insured + C(Vehicle_Age) + C(Vehicle_Damage) + Annual_Premium + Policy_Sales_Channel + Vintage'
logit_model = smf.logit(formula = str(f), data=df).fit()
logit_model.summary()
```
