Consider the inbuilt 'tips' dataset in seaborn package.

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips.head()
```

Use your choice of plotting library to plot the data to find answer to the following questions.

Q1) Which is the busiest week of the day?

```python
sns.countplot(x='day', data=tips)
plt.show()
```

Saturday is the busiest day of the week as more observations are made on this day

Q2) How does the total bill vary by sex?

```python
sns.boxplot(data=tips, x='sex', y='total_bill')
plt.show()
```

The range of total bill paid by Male customers is more than the Female customers.

Q3) Plot a box plot to see how Tip amount vary by day of the week?

```python
sns.boxplot(data=tips, x='tip', y='day')
plt.show()
```

More variation on the tips provided by the customer on a Sunday

Q4) Draw a plot to find out of smokers pay more or less than non-smokers on Weekends? How about Friday?

```python
sns.relplot(x="total_bill", y="tip", col="day",
          hue="smoker", data=tips)
plt.show()
```

1. On weekends, non-smokers pay more than smokers
2. On Friday, smokers pay more than non-smokers

Q5) Draw an appropriate plot to find out, between Male and Female, who tips higher on Dinner ? Is this any different during Lunch?

```python
sns.relplot(x="total_bill", y="tip", col="time",
          hue="sex", data=tips)
plt.show()
```

1. Male customers paid more tips than Female customers during Dinner.
2. But, this is not the same during Lunch.
