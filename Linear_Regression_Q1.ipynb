```python
# Import libraries
import numpy as np
import statsmodels.api as sm
```

```python
# Generate dataset
np.random.seed(2022)
n = 5          #starting value
x = np.random.uniform(0,10,100)
noise = np.random.randn(100)*n
y = 1.5*x + noise
y
```

```python
for i in range(5,51):
    # Generate dataset
    np.random.seed(2022)
    n = i          #starting value
    x = np.random.uniform(0,10,100)
    noise = np.random.randn(100)*n
    y = 1.5*x + noise    
    
    # Fit the model
    print(f"Iteration where n is {i}")
    x = sm.add_constant(x)             #adding the intercept term
    reg_model = sm.OLS(y,x).fit()      #fit an OLS model
    print(reg_model.summary())
    print()
```
