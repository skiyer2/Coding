```python
# Import libraries
import pandas as pd
from scipy.spatial import distance_matrix
```

```python
# Create dataframe
data = [[0.40,0.53], [0.22,0.38], [0.35,0.32], [0.26,0.19], [0.08,0.41], [0.45,0.30]]
points = ['Point P1', 'Point P2', 'Point P3', 'Point P4', 'Point P5', 'Point P6']

df = pd.DataFrame(data, columns=['X', 'Y'], index=points)
df
```

```python
# Create distance matrix
pd.DataFrame(distance_matrix(df.values, df.values), index=df.index, columns=df.index)
```
