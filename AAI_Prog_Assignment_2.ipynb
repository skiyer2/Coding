A) Numpy

Consider the code below:

```python
import numpy as np
np.random.seed(123)
```

```python
random_int = np.random.randint(1,high=1000,size=(20,5))
random_int
```

Qn 1) What is the average value of the first 5 rows ( 2 decimal place) ?

```python
avg_5_rows = round(random_int[:5].mean(),2)
avg_5_rows
```

Qn 2) What is the average value of the second element in each of the list (2 decimal places)?

```python
second_element = [random_int[i][1] for i in range(20)]
second_element_arr = np.array(second_element)
second_element_avg = round(second_element_arr.mean(),2)
second_element_avg
```

Q5) Write a python program to count the number occurrences of all vowels present in a string.

```python
def countvow(s):
    cnt = 0
    v = "aeiouAEIOU" #both lower and upper cases
    for l in s:
        if l in v:
            cnt+=1
    return cnt
```

```python
countvow("I am Karthik")
```

```python
countvow("I am practising python programming daily")
```
