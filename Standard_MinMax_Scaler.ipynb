```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import preprocessing

df = pd.DataFrame({'column1':np.concatenate([np.random.normal(10,2,1000), np.random.normal(1,2,20)]),
                   'column2':np.concatenate([np.random.normal(15,2,1000), np.random.normal(40,2,20)])})

scaler = preprocessing.StandardScaler()
standard_df = scaler.fit_transform(df)
standard_df = pd.DataFrame(standard_df, columns=['column1','column2'])

scaler = preprocessing.MinMaxScaler()
minmax_df = scaler.fit_transform(df)
minmax_df = pd.DataFrame(minmax_df, columns=['column1','column2'])
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols = 3, figsize =(20, 5))


sns.kdeplot(df['column1'], ax = ax1, color ='red')
sns.kdeplot(df['column2'], ax = ax1, color ='blue')
ax1.set_title('Before Scaling')

sns.kdeplot(standard_df['column1'], ax = ax2, color ='black')
sns.kdeplot(standard_df['column2'], ax = ax2, color ='g')
ax2.set_title('After Standard Scaling')
 
sns.kdeplot(minmax_df['column1'], ax = ax3, color ='black')
sns.kdeplot(minmax_df['column2'], ax = ax3, color ='g')
ax3.set_title('After Min-Max Scaling')

plt.show()
```
